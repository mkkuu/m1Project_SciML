# Tutoriel de Julia
Ce tutoriel a pour objectif de permettre à un utilisateur n'ayant jamais utilisé Julia de :
- installer Julia
- configurer un environnement de travail minimal
- comprendre les commandes basiques
- exécuter un script Julia
- et lancer le projet associé

## Installation de Julia
### Télécharger Julia
Julia est disponible gratuitement sur le site officiel :
[Julia](https://julialang.org/downloads/)

Ce projet utilise la version <b>1.10.3</b> , choississez la version correspondant à votre système (Windows, Linus, OS...)

Une fois l'installation terminée Julia peut être lancé :
- soit via le menu système
- soit via un terminal avec la commande julia

Il est fortement recommandé d'avoir également <b>Visual Studio Code</b> bien que Julia puisse être utilisé seul.

## Le REPL Julia : premier contact
Le REPL (Read-Eval-Print-Loop) est l'interface interactive de Julia.
Dans VS Code :
- ouvrir la palette de commandes avec *Ctrl + Shift + P*
- lancer Julia avec : 

```python
Start REPL
```

Vous devriez voir apparaître :

```python
julia>
```

## Premier programme : Hello World
Créer un fichier hello.jl
Ecrire println("Hello, World !")
Pour l'exécuter :
- soit sélectionner la ligne et appuyer sur *Crtl + Entrée*
- soit en lançant le fichier entier via :

```python
Run Julia File
```

## Variables et calculs de base
Julia est dynamiquement typé mais optimisé via l'interférence de types.
Code exemple :

```python
x = 3
y = 2.5
z = x + y
println(z)

# Code d'exemple de fonctions :
f(x) = x^2 + 2x + 1
println(f(3))

# Tableau et boucles
v = [1, 2, 3, 4]
println(v[1]) # l'indexation commence à 1
for i in 1:5
    println(i)
end
```

## Gestion des packages
Julia distingue l'installation d'un package et son chargement.

### Installer un package
Cette étape est à faire une seule fois par environnement :

```python
using Pkg
Pkg.add("Plots")
```

### Charger un package
Cette étape est à faire à chaque nouvelle session Julia
using Plots

## Exécuter un script Julia
Dans VS Code il existe deux méthodes :
### Mode interactif (Ctrl + Entrée)
- envoie la ligne ou la séléction dans le REPL
- l'état dépend de ce qui a déjà été exécuté
- pratique pour tester mais peut provoquer des erreurs si l'ordre n'est pas respecté

### Exécuter tout le fichier
- utiliser :

```python
Run Julia File
```

- Le script est exécuté dans son intégralité
- recommandé pour vérifier que le code est reproductible

## Raccourcis utiles dans VS Code
- Lancer le REPL : *Ctrl + Shift + P*
- Envoyer une ligne / sélection : *Ctrl + Entrée*
- Envoyer un bloc : *Alt + Entrée*
- Exécuter le fichier entier : 

```python
Run Julia File
```

- Afficher le Plot Viewer : 

```python
Julia: Show Plot Viewer
```

## Exemple de référence : tracer une courbe

```python
using Plots

x = 0:0.1:10
y = sin.(x)

plot(x, y, title="Exemple de tracé en Julia")
```

## Pour lancer un projet
1. Ouvrir le dossier du projet dans le VS Code
2. Démarrer le REPL Julia
3. Installer les dépendances si nécessaire :

```python
using Pkg
Pkg.instantiate()
```

4. Exécuter le script principal du projet
