```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Projection
proj = ccrs.PlateCarree()

fig = plt.figure(figsize=(6,5))
ax = plt.axes(projection=proj)

# Domaine Manche 
ax.set_extent([-5, 2, 48, 51], crs=proj)

# Côtes
ax.add_feature(cfeature.COASTLINE, linewidth=1)

# Fleuves
# Liste des noms 
river_names = ["Seine", "Somme", "Trieux", "Rance"]

# Longitudes 
river_lons = [0.1, 1.5, -3.0, -2.0]

# Latitudes
river_lats = [49.5, 50.2, 48.8, 48.6]


# On parcourt les indices de la liste (0, 1, 2, 3)
for i in range(len(river_names)):

    name = river_names[i]
    lon = river_lons[i]
    lat = river_lats[i]

    # On trace le point 
    ax.plot(
        lon, lat,
        marker="o",
        markersize=4,
        color="blue",
        transform=proj
    )

    # On écrit le nom 
    ax.text(
        lon + 0.1,
        lat + 0.05,
        name,
        fontsize=8,
        transform=proj
    )

ax.plot(
    -1.3,50.8,marker="o",markersize =4, 
    color="cyan",transform=proj)
ax.text(
    -1.2,50.85,"Solent",fontsize=8,
    transform=proj
)
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='white')

ax.set_title("Carte de contexte géographique de la Manche")

plt.tight_layout()
plt.savefig("../docs/referenceMap.png", dpi=300)
plt.show()
```
