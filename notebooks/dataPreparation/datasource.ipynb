# Source and explanations of the pretraitement for our datas

## Data from Copernicus

### Presentation of the product SST_GLO_SST_L4_REP_OBSERVATIONS_010_024

We chose our first set of data from the site Copernicus more exactly the set of data [C3S-GLO-SST-L4-REP-OBS-SST](https://data.marine.copernicus.eu/product/SST_GLO_SST_L4_REP_OBSERVATIONS_010_024/download?dataset=C3S-GLO-SST-L4-REP-OBS-SST_202506) from 2010 to 2020.

All of our data comes from the Copernicus Marine Environment Monitoring Service (CMEMS). This is a program operational of the European Comission dedicated to the observation and the modelisation of the oceans.

We chose the product type "L4" which refer to the level of processing of our dataset, the level 4 being a combination of both the level 2 and 3 plus a serie of processes then an Optimal Interpolation methodology obtain through a multi-captors observation.

All the observations from the satellites are submitted to controls of quality then combined using an algorithm of Optimal Analyse (OSTIA) wich permits to produce continuous fields of SST of a regular grid of 0.05°.

A high resolution land mask is used which ensure that we don't have any land data in our dataset[ProductUserManual] (https://documentation.marine.copernicus.eu/PUM/CMEMS-SST-PUM-010-024.pdf).

Concerning the [QualityInformationDocument] (https://documentation.marine.copernicus.eu/QUID/CMEMS-SST-QUID-010-024.pdf?utm_source=chatgpt.com) the quality of the product is evaluated and documented in the Quality Information Document which an estimation of the errors associated with the product.

## Data from NOAA

### Presentation of the product NOAA OISST v2.1

We used the dataset [NOAA_OI_SST_V2_High_Resolution] (https://psl.noaa.gov/data/gridded/data.noaa.oisst.v2.highres.html) from 2010 to 2020.

The data used come from the National Oceanic and Atmospheric Administration (NOAA) which is an federal american agency responsable for the production and the diffusion of climatic and oceanic datas.

The product NOAA OISST v2.1 refers to a field of SST analysed daily by the use of a Optimum Interpolation method based on satellite observations. This place this product to a level of processing of our dataset equal to the Level 4 ("L4").
By reading the Climate Data Record [CDR](https://www.ncei.noaa.gov/pub/data/sds/cdr/CDRs/Sea_Surface_Temperature_Optimum_Interpolation/AlgorithmDescription_01B-09.pdf?utm_source=chatgpt.com) Program and more specifically the Climate Algorithm Theoretical Basis Document (C-ATBD) we can read in details the chain of processing.
The chain of processing is at it follow :
- Collect of the datas from the AVHRR satellites and other sources (buoy, ships and Argo)
- Quality control by the filtering of aberrant datas
- Correction of the bias by adjusting the satellite
- Interpolation optimum by using a wide radius correlation and the production of a global field constant of 0.25°

The product OISST v2.1 belongs to the Climate Data Records (CDR) from NOAA which ensure that we have a temporal stability, a traceability complete of the algorithm and an independant validation by comparaison with in situ observations.
